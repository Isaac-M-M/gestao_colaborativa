```python
Pseudocódigo  Sistema Gestão Colaborativa (fluxo principal)

Aqui eu fiz como se fosse um sistema de linha de comando que representa o que o Django faz “por trás” da interface.

INÍCIO

    // Carrega página inicial (lista de projetos)
    ENQUANTO VERDADEIRO FAÇA

        LIMPAR_TELA()
        ESCREVER "===== GESTÃO COLABORATIVA ====="
        ESCREVER "Carregando lista de projetos..."

        projetos ← BUSCAR_TODOS_PROJETOS()

        SE projetos ESTÁ_VAZIO ENTÃO
            ESCREVER "Nenhum projeto cadastrado."
        SENÃO
            ESCREVER "ID | TÍTULO | RESPONSÁVEL | STATUS"
            PARA CADA projeto EM projetos FAÇA
                ESCREVER projeto.id, projeto.titulo, projeto.responsavel, projeto.status
            FIMPARA
        FIMSE

        ESCREVER ""
        ESCREVER "Opções:"
        ESCREVER "1 - Criar novo projeto"
        ESCREVER "2 - Ver detalhes de um projeto"
        ESCREVER "3 - Editar um projeto"
        ESCREVER "4 - Excluir um projeto"
        ESCREVER "0 - Sair"
        ESCREVER ""
        LER opcao

        SE opcao = 1 ENTÃO
            // Criar projeto
            LIMPAR_TELA()
            ESCREVER "=== Novo Projeto ==="
            ESCREVER "Informe os dados do projeto:"

            LER titulo
            LER descricao
            LER prazo
            LER responsavel
            LER status

            SE titulo = "" OU responsavel = "" ENTÃO
                ESCREVER "Erro: Título e responsável são obrigatórios."
                PAUSAR()
            SENÃO
                novo_projeto ← CRIAR_PROJETO(titulo, descricao, prazo, responsavel, status)
                SALVAR_PROJETO(novo_projeto)
                ESCREVER "Projeto criado com sucesso!"
                PAUSAR()
            FIMSE

        SENÃO SE opcao = 2 ENTÃO
            // Detalhes do projeto + tarefas + comentários
            LIMPAR_TELA()
            ESCREVER "Digite o ID do projeto para ver detalhes:"
            LER id_projeto

            projeto ← BUSCAR_PROJETO_POR_ID(id_projeto)

            SE projeto = NULO ENTÃO
                ESCREVER "Projeto não encontrado."
                PAUSAR()
            SENÃO
                tarefas ← BUSCAR_TAREFAS_DO_PROJETO(projeto)
                comentarios ← BUSCAR_COMENTARIOS_DO_PROJETO(projeto)

                ESCREVER "=== Detalhes do Projeto ==="
                ESCREVER "Título: ", projeto.titulo
                ESCREVER "Descrição: ", projeto.descricao
                ESCREVER "Prazo: ", projeto.prazo
                ESCREVER "Responsável: ", projeto.responsavel
                ESCREVER "Status: ", projeto.status

                ESCREVER ""
                ESCREVER "Tarefas:"
                SE tarefas ESTÁ_VAZIO ENTÃO
                    ESCREVER "Nenhuma tarefa cadastrada."
                SENÃO
                    PARA CADA tarefa EM tarefas FAÇA
                        ESCREVER "- ", tarefa.titulo, " [", tarefa.status, "]"
                    FIMPARA
                FIMSE

                ESCREVER ""
                ESCREVER "Comentários:"
                SE comentarios ESTÁ_VAZIO ENTÃO
                    ESCREVER "Nenhum comentário."
                SENÃO
                    PARA CADA c EM comentarios FAÇA
                        ESCREVER "- ", c.autor, ": ", c.texto
                    FIMPARA
                FIMSE

                ESCREVER ""
                ESCREVER "Ações rápidas:"
                ESCREVER "1 - Adicionar tarefa"
                ESCREVER "2 - Adicionar comentário"
                ESCREVER "Qualquer outra tecla - Voltar"
                LER subopcao

                SE subopcao = 1 ENTÃO
                    // fluxo resumido de cadastro de tarefa
                    LER titulo_tarefa
                    LER descricao_tarefa
                    LER responsavel_tarefa
                    CRIAR_E_SALVAR_TAREFA(projeto, titulo_tarefa, descricao_tarefa, responsavel_tarefa)
                    ESCREVER "Tarefa criada com sucesso!"
                    PAUSAR()
                SENÃO SE subopcao = 2 ENTÃO
                    LER texto_comentario
                    LER anexo_opcional
                    CRIAR_E_SALVAR_COMENTARIO(projeto, texto_comentario, anexo_opcional)
                    ESCREVER "Comentário adicionado com sucesso!"
                    PAUSAR()
                SENÃO
                    // Voltar ao menu
                FIMSE

            FIMSE

        SENÃO SE opcao = 3 ENTÃO
            // Editar projeto
            LIMPAR_TELA()
            ESCREVER "Digite o ID do projeto para editar:"
            LER id_projeto
            projeto ← BUSCAR_PROJETO_POR_ID(id_projeto)

            SE projeto = NULO ENTÃO
                ESCREVER "Projeto não encontrado."
                PAUSAR()
            SENÃO
                ESCREVER "Deixe em branco para manter o valor atual."

                ESCREVER "Novo título (atual: ", projeto.titulo, "):"
                LER novo_titulo
                ESCREVER "Nova descrição (atual: ", projeto.descricao, "):"
                LER nova_descricao
                ESCREVER "Novo prazo (atual: ", projeto.prazo, "):"
                LER novo_prazo
                ESCREVER "Novo responsável (atual: ", projeto.responsavel, "):"
                LER novo_responsavel
                ESCREVER "Novo status (atual: ", projeto.status, "):"
                LER novo_status

                ATUALIZAR_PROJETO(projeto, novo_titulo, nova_descricao, novo_prazo, novo_responsavel, novo_status)
                SALVAR_PROJETO(projeto)

                ESCREVER "Projeto atualizado com sucesso."
                PAUSAR()
            FIMSE

        SENÃO SE opcao = 4 ENTÃO
            // Excluir projeto
            LIMPAR_TELA()
            ESCREVER "Digite o ID do projeto para excluir:"
            LER id_projeto
            projeto ← BUSCAR_PROJETO_POR_ID(id_projeto)

            SE projeto = NULO ENTÃO
                ESCREVER "Projeto não encontrado."
                PAUSAR()
            SENÃO
                ESCREVER "Tem certeza que deseja excluir o projeto? (S/N)"
                LER confirma

                SE confirma = "S" OU confirma = "s" ENTÃO
                    EXCLUIR_PROJETO(projeto)
                    ESCREVER "Projeto excluído com sucesso."
                SENÃO
                    ESCREVER "Exclusão cancelada."
                FIMSE
                PAUSAR()
            FIMSE

        SENÃO SE opcao = 0 ENTÃO
            // Sair do sistema
            ESCREVER "Encerrando a Gestão Colaborativa..."
            SAIR_DO_LACO ← VERDADEIRO
            SE SAIR_DO_LACO = VERDADEIRO ENTÃO
                QUEBRAR // sai do ENQUANTO
            FIMSE

        SENÃO
            ESCREVER "Opção inválida."
            PAUSAR()
        FIMSE

    FIMENQUANTO

FIM


Esse pseudocódigo representa bem o que o projeto faz com Django + DRF + FastAPI, só que em forma de lógica pura.
```
